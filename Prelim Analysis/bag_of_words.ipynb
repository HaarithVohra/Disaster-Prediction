```python
dataRelevant = [data.text[i] for i in range(len(data)) if data.choose_one[i] == 1]
```

```python
for i in range(10):
    print(dataRelevant[i])
```

```python
# Design the vocab
count_vectorizer = CountVectorizer()

# Create bag 
bag_of_words = count_vectorizer.fit_transform(dataRelevant)

# Show bag_of_words model as DataFrame
feature_names = count_vectorizer.get_feature_names()
bag_of_words_vector = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
bag_of_words_vector
```

```python
word_list = []
word_freq = []
for i in range(10):
    save_max = []
    for i in range(len(bag_of_words_vector)):
        col_names = bag_of_words_vector.columns[i]
        col_vals = bag_of_words_vector[col_names].tolist()
        save_max.append(sum(col_vals))

    x = save_max.index(max(save_max))
    print(max(save_max))
    word_freq.append(max(save_max))
    print(x)

    col_name = bag_of_words_vector.columns[x]
    word_list.append(bag_of_words_vector.columns[x])
    print(col_name)
    # max_list = []
    # for i in range(10):
    #     x = save_max.index(max(save_max))
    #     occur = max(save_max)
    #     col_name = bag_of_words_vector.columns[x]
    bag_of_words_vector = bag_of_words_vector.drop(col_name, axis=1)
#     max_list.append(col_name)

# print(max_list)

```

```python
label = word_list
values = word_freq
print (word_list)
print(word_freq)
def plot_bar():
    # this is for plotting purpose
    index = np.arange(len(label))
    plt.bar(index, values)
    plt.xlabel('Words (Features)', fontsize=7)
    plt.ylabel('Frequency', fontsize=10)
    plt.xticks(index, label, fontsize=10, rotation=30)
    plt.title('Most Significant Features')
    plt.show()
plot_bar()
```
